```python
import matplotlib.pyplot as plt
print(plt. __version__)
```

```python
proteincount={'A': 4,
  'C': 2,
  'D': 2,
  'E': 2,
  'F': 2,
  'G': 4,
  'H': 2,
  'I': 3,
  'K': 2,
  'L': 6,
  'M': 1,
  'N': 2,
  'P': 4,
  'Q': 2,
  'R': 6,
  'S': 6,
  'T': 4,
  'V': 4,
  'W': 1,
  'Y': 2,
  '*': 3}
```

```python
proteinname=list(proteincount.keys())
proteinc=list(proteincount.values())
```

## Bar Graph

```python
plt.bar(proteinname, proteinc, color ='maroon')
plt.xlabel("Protein")
plt.ylabel("Protein Count")
plt.title("Protein Visualization")
plt.show()
```

## Pie Chart

```python
plt.pie(proteinc, labels = proteinname, startangle = 180)
plt.title("Protein Visualization")
plt.show()
```

## Line graph

```python
plt.plot(proteinname,proteinc,color='maroon')
plt.xlabel("Protein")
plt.ylabel("Protein Count")
plt.title("Protein Visualization")
plt.show()
```

## Function

```python
def visualizeProtein(proteincount,plottype,graphcolor='maroon',dpival=100):
    """
    Pass a dictionary of protein_count will plot a graphs based upon graphtype('line' or 'bar' or 'pie').
    Change color by setting graphcolor='colorcode/name', By default = 'maroon'
    Also Change size of graph by setting dpival= integer_value, By default=100
    
    Input: Dictionary of Protein Count(required), plottype(required), graphcolor(optional), dpival(optional)
    Output: Matplotlib Graph object
    """
    import matplotlib.pyplot as plt
    
    proteinname=list(proteincount.keys())
    proteinc=list(proteincount.values())
    
    plt.figure(dpi=dpival)
    
    if(plottype=='bar'):
        plt.bar(proteinname, proteinc, color =graphcolor)
        plt.xlabel("Protein")
        plt.ylabel("Protein Count")
        plt.title("Protein Visualization")
        return plt
    elif(plottype=='pie'):
        plt.pie(proteinc, labels = proteinname, startangle = 180)
        plt.title("Protein Visualization")
        return plt
    elif(plottype=='line'):
        plt.plot(proteinname,proteinc,color=graphcolor)
        plt.xlabel("Protein")
        plt.ylabel("Protein Count")
        plt.title("Protein Visualization")
        return plt
    else:
        return "Available Graphs ['bar','line','pie']"
    
```

## Testing

```python
visualizeProtein(proteincount,'bar')
```

```python
visualizeProtein(proteincount,'line',graphcolor='green')
```
